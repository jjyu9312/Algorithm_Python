# K번째 수 구하기

```python
''' 예시 '''

''' 
2 
6 2 5 3
5 2 7 3 8 9 
15 3 10 3
4 15 8 16 6 6 17 3 10 11 18 7 14 7 15
'''
```

```python
T = int(input('테이블 입력 : '))

for t in range(T):
    n, s, e, k = map(int, input().split())
    a = list(map(int, input().split()))
    a = a[s-1:e+1]
    a.sort()
    print(f"{t}테이블의 k번째 숫자는 {a[k-1]}") 
```

# K번째 큰 수 구하기

```python
# 예시
# list 중 3장을 뽑아 각 카드에 적힌 수를 합한 값을 기록 후 K번째 큰 수를 구함
'''
10 3
13 15 34 23 45 65 33 11 26 42

답 : 143 
'''
```

```python
# 내 방법

t, n = map(int, input().split())
a = list(map(int, input().split()))

k = set(a)
k = sorted(k)

print(k)
print(f"{n}번째 큰 수는 {k[n-1]}")
```

```python
# 강의 방법

n, k = map(int, input().split())
a = list(map(int, input().split()))
res = set()

# 3장을 뽑는 과정
for i in range(n):
    for j in range(i+1, n):
        for m in range(j+1, n):
            res.add(a[i] + a[j] + a[m])

res = list(res)
res.sort(reverse = True)

print()
print(res)
print()
print(res[k-1])
```
